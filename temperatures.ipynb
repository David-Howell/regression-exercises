```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
f = np.arange(0,1000)
f
```

```python
c = (f - 32) / 1.8
c
```

```python
temps = pd.DataFrame(data=[c, f], columns=np.arange(0,1000), index=['C', 'F']).T
```

```python
plt.figure(figsize=(8,8))
sns.scatterplot(data=temps, y='C', x='F')
plt.xlim(-100,1000)
plt.ylim(-100,1000)
```
